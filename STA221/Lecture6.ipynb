K-MEANS CLUSTERING ON HALF-MOON DATASET

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances_argmin

from sklearn.datasets import make_moons
X, y = make_moons(200, noise=.05, random_state=0)
```

```python
labels = KMeans(2, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

FIXING THIS ISSUE WITH SPECTRAL CLUSTERING

```python
from sklearn.cluster import SpectralClustering

#model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', assign_labels='kmeans')

model = SpectralClustering(n_clusters=2, affinity='rbf', gamma=0.5, assign_labels='kmeans')

labels = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

K-MEANS ON REAL_WORLD (DIGITS) DATASET

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape ## 1,797 samples with 64 features (pixels in an 8×8 image)
```

```python
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8) ## RESHAPE 64 dimensional vectors into 8x8 images for visualization
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

```python
from sklearn.cluster import SpectralClustering
model = SpectralClustering(n_clusters=10, affinity='nearest_neighbors',
                           assign_labels='kmeans')
labels = model.fit_predict(digits.data)
print(labels.size)
```
